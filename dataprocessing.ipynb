```python
import pandas as pd
```

```python
df_colleges = pd.read_csv('collegesranked.csv')
df_colleges.head()
```

```python
df_colleges_melted = pd.melt(df_colleges, id_vars=['Name', 'State', 'Conference', 'PrivatePublic'], var_name='Year', value_name='Rank')
df_colleges_melted = df_colleges_melted.dropna(subset=['Rank'])
df_colleges_melted['Rank'] = df_colleges_melted['Rank'].astype(int)
df_colleges_melted
```

```python
df_colleges_melted.to_csv('collegesformatted.csv')
```
